```python
import pandas as pd
df= pd.read_excel('products.xlsx')
df
```

```python
# Remove duplicate rows
df = df.drop_duplicates()
df
```

```python
# Drop rows where any column has a missing value
df = df.dropna()
df
```

```python
#splitting the Employee Address which contain Employee Address and their city seprated by a comma(,)
df[['EmpAddress','EmpCity']]=df['Employee Address'].str.split(',',expand=True)
```

```python
df[['EmpAddress','EmpCity']]
```

```python
df[['EmpFName','EmpLName']]=df['Employee'].str.split(' ',expand=True)
```

```python
df[['EmpPhone']] =df[['EmployeePhone']]
df[['EmpPhone']]
```

```python
df[['EmpFName','EmpLName']]
```

```python
# Employee CSV
Employee_csv=df[['EmpFName','EmpLName','EmpAddress','EmpCity','EmpPhone']]
Employee_csv.drop_duplicates(subset=['EmpFName', 'EmpLName']) 
```

```python
Employee_csv.drop_duplicates(subset=['EmpFName', 'EmpLName']).to_excel('Employee.xlsx', index=False) 
```

```python
df[['DiscountPerUnit']]=df[['Discount']]
df[['DiscountPerUnit']]
```
